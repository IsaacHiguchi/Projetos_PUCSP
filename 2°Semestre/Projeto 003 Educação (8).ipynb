```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

def obter_driver():
    return webdriver.Chrome()

def abrir_pagina(driver):
    for pag in range(1):
        driver.get('https://en.wikipedia.org/wiki/List_of_countries_by_literacy_rate')   
```

```python
driver = obter_driver()
abrir_pagina(driver)
```

```python
import pandas as pd

country = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[1]/a')
literate_rate = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[2]')



rate = []
for i in range(len(country)):
    data = {'Country':country[i].text,
            'Literate_rate':literate_rate[i].text}

    rate.append(data)
    
df = pd.DataFrame(rate)
df
```

```python
df['Literate_rate'] = df['Literate_rate'].str.replace("%", "")
df['Literate_rate'] = df['Literate_rate'].str.replace(",", ".")
df['Literate_rate'] = df['Literate_rate'].str.replace("not reported by UNESCO 2015", "N/A")
df
```

```python
#lista de paises que possuem dados, porem nao é possivel converter para numero

l1 = df.loc[df['Country'] == 'Bangladesh']
print(l1)
l2 = df.loc[df['Country'] == 'India']
print(l2)
l3 = df.loc[df['Country'] == 'Jordan']
print(l3)
l4 = df.loc[df['Country'] == 'Morocco']
print(l4)
l5 = df.loc[df['Country'] == 'Myanmar']
print(l5)
l6 = df.loc[df['Country'] == 'Pakistan']
print(l6)
```

```python
import pandas as pd
rend_pc = pd.read_csv('Rendimento_PC-2019.csv', sep = ";")
rend_2019pc= rend_pc[['Abertura geográfica', 'ind', 'classe', 'sub.classe', 'Unid', '2019']]
rend_2019pc
```

```python
rend_2019pc = rend_2019pc.rename(columns={'2019': 'ANO2019'})
rend_2019pc
```

```python
rend_2019pc['ANO2019'] = rend_2019pc.ANO2019.astype(float)
rend_2019pc
```

```python
#Total de rendimentos per capita

total_rendpc = rend_2019pc['ANO2019'].sum()
total_rendpc
```

```python
#Rendimento per capita da região Norte do Brasil

norte_rendpc = rend_2019pc.iloc[0]['ANO2019']
norte_rendpc
```

```python
#O quanto a região Norte representa do total de rendimentos per capita

porcentagem_n = round(((norte_rendpc*100)/total_rendpc), 2)
porcentagem_n
```

```python
#Rendimento per capita da região Nordeste do Brasil

nordeste_rendpc = rend_2019pc.iloc[1]['ANO2019']
nordeste_rendpc
```

```python
#O quanto a região Sudeste representa do total de rendimentos per capita

porcentagem_se = round(((sudeste_rendpc*100)/total_rendpc), 2)
porcentagem_se
```

```python
#Rendimento per capita da região Sul do Brasil

sul_rendpc = rend_2019pc.iloc[3]['ANO2019']
sul_rendpc
```

```python
#O quanto a região Sul representa do total de rendimentos per capita

porcentagem_s = round(((sul_rendpc*100)/total_rendpc), 2)
porcentagem_s
```

```python
#O quanto as regiões Sul e Sudeste representam do total de rendimentos per capita

porcentagem_s_se = round(((sul_sudeste_rendpc*100)/total_rendpc), 2)
porcentagem_s_se
```
