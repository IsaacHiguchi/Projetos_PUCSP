```python
class ExtratorListaUber:
    """
    O ExtratorUEPGVest é uma classe utilitária que facilita a extração das informações sobre o vestibular atual da Univesidade Estadual de Ponta Grossa (UEPG)   
    """
    def __init__(self, driver=None):
        """
        Construtor
        """
        self.driver = webdriver.Chrome() if not driver else driver
        self.driver.get("https://www.uber.com/eligible-vehicles/")
        self.lst_carros = ''
        self.qntd_linhas = 265
        
    def carros(self):
        for i in range (1, self.qntd_linhas+1):    
            xpath_marcas = f'//*[@id="vehicles"]/div/div[2]/div/div/div/ul/li[{i}]'
            marcas = self.driver.find_element_by_xpath(xpath_marcas)
            self.lst_carros += (marcas.text) + '\n'
            marcas.click()
            j = 1
            while j < 1000:
                xpath_carros = f'//*[@id="vehicles"]/div/div[2]/div/div/div/ul/li[1]/div[2]/div/div/div/div/div/ul/li[{j}]'
                try:
                    carros = self.driver.find_element_by_xpath(xpath_carros)
                    self.lst_carros += (carros.text) + '\n'
                    j += 1
                except:
                    break
```

```python
uber = ExtratorListaUber()
```

```python
uber.carros()
```
