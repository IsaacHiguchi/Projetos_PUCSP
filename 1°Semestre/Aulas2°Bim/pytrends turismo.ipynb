```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='pt-br')  

keywords = ["passeio em", "cancelar passagem"]
pytrends.build_payload(keywords, timeframe='today 5-y')
```

```python
data = pytrends.interest_over_time()
data = data.drop('isPartial', axis=1)

data.tail()
```

```python
import matplotlib.pyplot as plt

plt.plot(data)

plt.suptitle('Pesquisas no Google')
plt.xlabel('anos')                       
plt.ylabel('pesquisas semanais')  


plt.legend(keywords, loc='upper left')

plt.show()
plt.savefig('googletrendsturismo.png')
```
